```python
avg_rating = game_df.groupby("category")["average rating"].mean().sort_values(ascending=False)
print(avg_rating)
# game_df.groupby("category") -> Groups all games by their category.
# ["total ratings"] -> Selects the 'total ratings' column because that's what we want to analyse.
# .mean() -> Calculates the average rating for each category.
# .sort_values(ascending=False) -> Sorts the results from highest to lowest, so the best-rated categories appear at the top.
```

```python
avg_rating.plot(kind="barh", color="lightblue", figsize=(10,6))
plt.ylabel("Average Rating")
plt.title("Average Rating per Game Category")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,6))

categorical_count = game_df['price'].value_counts()
axes[0].bar(x=categorical_count.index, height=categorical_count)
axes[0].set_title("Price_Game")

categorical_count = game_df.value_counts('category',normalize=True)
axes[1].pie(x=categorical_count, labels=categorical_count.index)
axes[1].set_title("Game_Category")

axes[2].scatter(data=game_df, x='5 star ratings', y= '1 star ratings')
axes[2].set_title("Rating")

plt.tight_layout()
plt.show()
```
